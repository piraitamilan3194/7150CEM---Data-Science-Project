```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
maintainance = pd.read_csv("Prediction.csv")
maintainance.head()
```

```python
maintainance.describe()
```

```python
maintainance.corr()
```

```python
X = maintainance[['Lub oil pressure', 'Fuel pressure', 'Coolant pressure', 'lub oil temp']]
Y = maintainance['prediction(Engine rpm)']
```

```python
X = sm.add_constant(X)
```

```python
reg_res = sm.OLS(Y,X).fit()
```

```python
reg_res.summary()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_Test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.3, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, Y_train)
```

```python
C = lr.intercept_
C
```

```python
M = lr.coef_
M
```

```python
Y_pred_train = lr.predict(X_train)
Y_pred_train
```

```python
plt.scatter(Y_train, Y_pred_train)
plt.xlabel("Actual Engine Rpm")
plt.ylabel("Predicted Engine Rpm")
plt.show()
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(Y_train, Y_pred_train)
```

```python
Y_pred_test = lr.predict(X_Test)
Y_pred_test
```

```python
plt.scatter(Y_test, Y_pred_test)
plt.xlabel("Actual Engine Rpm")
plt.ylabel("Predicted Engine Rpm")
plt.show()
```

```python
r2_score(Y_test, Y_pred_test)
```
